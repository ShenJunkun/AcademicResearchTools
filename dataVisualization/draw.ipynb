### 绘制柱状图

#### 一个x对应一个y

```python
import matplotlib.pyplot as plt
import matplotlib

# %matplotlib inline
# 柱状图数据
x_values = ['A', 'B', 'C', 'D']
y_values = [20, 35, 30, 25]
# plt.rcParams['font.family'] = 'SimHei'

# 创建柱状图
plt.bar(x_values, y_values,label="legend")

# 设置图表标题和横纵坐标标签
plt.title("title")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
# 显示柱状图
plt.show()
```

#### 一个x对应两个y

```python
import numpy as np  
import matplotlib.pyplot as plt

#柱状图数据
x_values = [0.1, 0.2, 0.3, 0.4]
y1_values = [20, 35, 30, 25]
y2_values = [10, 25, 20, 15]

# 设置柱子宽度
bar_width = 0.35

#计算两组数据柱子的位置
bar1_positions = np.arange(len(x_values))
print(bar1_positions)
bar2_positions = bar1_positions + bar_width
print(bar2_positions)

#创建柱状图
plt.bar(bar1_positions, y1_values, width=bar_width, label="time")
plt.bar(bar2_positions, y2_values, width=bar_width, label="mem")

#设置图标标题和横纵坐标标签
plt.title("diagram")
plt.xlabel("radio")

#设置图例
plt.legend()

#显示柱状图
plt.show()
```

#### 一个x,2个y值，并在左右y轴表示刻度

```python
import numpy as np
import matplotlib.pyplot as plt

# 柱状图数据
x_values = [0.1, 0.3, 0.5, 0.7, 0.9]
y1_values = [165, 111, 111, 103, 102]
y2_values = [2876.29, 3388.29, 3388.29, 6460.29, 6460.29]

bar_width = 0.35  # 设置柱子宽度

# 计算两组数据柱子的位置
bar1_positions = np.arange(len(x_values))
bar2_positions = bar1_positions + bar_width

# 创建主坐标轴
fig, ax1 =plt.subplots()

# 创建次坐标轴
ax2 = ax1.twinx()

#绘制Y1数据
ax1.bar(bar1_positions, y1_values, width=bar_width, label='time')
ax1.set_ylabel('time')

#绘制Y2数据
ax2.bar(bar2_positions, y2_values, width=bar_width, label="mem", color='green')
ax2.set_ylabel('max mem')

#设置图标标题和横坐标标签
ax1.set_title("mem threshold")
ax1.set_xlabel("radio")

#合并图例
handles, labels = ax1.get_legend_handles_labels()
handles2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(handles + handles2, labels + labels2, loc='upper center')

#显示柱状图
plt.show()
```
